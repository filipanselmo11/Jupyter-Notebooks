```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Criando uma lista de idades ficticias
idades = [25, 30, 35, 40, 45, 30, 35, 40, 25, 30, 35, 60, 7, 25, 25, 30, 20, 10, 10, 11, 11, 11]
```

```python
#Convertendo a lista em uma serie do pandas
idades_series = pd.Series(idades)
```

```python
#Calculando a distribuicao de frequencia 
distribuicao_frequencia_idades = idades_series.value_counts().sort_index()
```

```python
print("Distribuicao de Frequencia de Idades:")
print(distribuicao_frequencia_idades)
```

```python
#Criando um histograma a partir da distribuicao de frequencia
plt.bar(distribuicao_frequencia_idades.index, distribuicao_frequencia_idades.values, color='skyblue')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.title('Histograma das Idades')
plt.show()
```

```python
#Criando um grfico de linhas 
plt.plot(distribuicao_frequencia_idades.index, distribuicao_frequencia_idades.values, marker='o', linestyle='--')
plt.xlabel('Idade')
plt.ylabel('Frequncia')
plt.title('Grfico de linhas das Idades')
plt.show()
```

```python
#Criando o grfico de colunas
plt.bar(distribuicao_frequencia_idades.index, distribuicao_frequencia_idades.values, color='lightgreen', width=0.5)
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.title('Grfico de Colunas das Idades')
plt.show()
```

```python
plt.pie(distribuicao_frequencia_idades, labels=distribuicao_frequencia_idades.index, autopct='%1.1f%%', startangle=140)
plt.title('Grfico de setores das Idades')
plt.axis('equal')
plt.show()
```
