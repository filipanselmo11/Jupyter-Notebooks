a = Coeficiente Angular
n = Nmero de elementos
Y = Moscow
X = datas

a = n * Soma(Xi - Yi) - Soma(Xi) * Soma(Yi) / n * Soma(Xi**2) - (Soma(Xi)**2)

```python
Y = moscow
X = datas
n = np.size(moscow)
```

```python
X = X**2
```

```python
X.shape
```

```python
#Coeficiente Angular
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y)) / (n*np.sum(X**2) - np.sum(X)**2)
a
```

```python
#Coeficiente linear
b = np.mean(Y) - a*np.mean(X)
b
```

```python
y = a*X+b
y
```

```python
np.linalg.norm(moscow-y)
```

```python
plt.plot(datas, moscow)
plt.plot(x,y)
plt.plot(41.5, 41.5*a+b, '*r')
```

```python
#Estimando valores para o futuro
plt.plot(datas, moscow)
plt.plot(x,y)
plt.plot(100, 100*a+b, '*r')
```
