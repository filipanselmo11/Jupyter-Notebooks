```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from prophet import Prophet
```

```python
dados = yf.download('JNJ', start='2020-01-01', end='2023-12-31', progress=False)
dados = dados.reset_index()
dados
```

```python
dados_treino = dados[dados['Date'] < '2023-07-31']
dados_teste = dados[dados['Date'] >= '2023-07-31']
```

```python
dados_pro_treino = dados_treino[['Date', 'Close']].rename(columns={'Date':'ds', 'Close':'y'})
dados_pro_treino
```

```python
modelo = Prophet(weekly_seasonality=True, yearly_seasonality=True, daily_seasonality=False)
modelo.add_country_holidays(country_name='US')
modelo.fit(dados_pro_treino)
```

```python
# Criar datas futuras para previsao atÃ© o final de 2023
futuro = modelo.make_future_dataframe(periods=150)
previsao = modelo.predict(futuro)
```

```python
previsao
```

```python
plt.figure(figsize=(14,8))
plt.plot(dados_treino['Date'], dados_treino['Close'], label='Dados de Treino', color='blue')
plt.plot(dados_teste['Date'], dados_teste['Close'], label='Dados Reais', color='green')
plt.plot(previsao['ds'], previsao['yhat'], label='PrevisÃ£o', color='orange', linestyle='--')

plt.axvline(dados_treino['Date'].max(), color='red', linestyle='--', label='Inicio da PrevisÃ£o')
plt.xlabel('Data')
plt.ylabel('PreÃ§o de Fechamento')
plt.title('PrevisÃ£o do PreÃ§o de Fechamento VS Dados Reais')
plt.legend()
plt.show()
```
