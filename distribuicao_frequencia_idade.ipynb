```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lista de idades (exemplo)
idades = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
```

```python
# Criar um DataFrame do Pandas com os dados de idade
df = pd.DataFrame(idades, columns=['Idade'])
```

```python
# Calcular a contagem de cada idade
contagem_idades = df['Idade'].value_counts().sort_index()
contagem_idades
```

```python
# Calcular a frequência relativa
frequencia_relativa = contagem_idades / contagem_idades.sum()
```

```python
# Calcular a frequência relativa acumulada
frequencia_relativa_acumulada = frequencia_relativa.cumsum()
```

```python
# Criar um DataFrame para a distribuição de frequência
distribuicao_frequencia_idade = pd.DataFrame({
    'Idade': contagem_idades.index,
    'Frequencia': contagem_idades.values,
    'Frequncia_Relativa': frequencia_relativa.values,
    'Frequencia_Relativa_Acumulada': frequencia_relativa_acumulada.values
})
distribuicao_frequencia_idade
```

```python
# Histograma
plt.hist(df['Idade'], bins=len(contagem_idades), edgecolor='black')
plt.title('Histograma das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

```python
# Gráfico de barras
plt.bar(contagem_idades.index, contagem_idades.values, color='skyblue')
plt.title('Gráfico de Barras das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.grid(axis='y')
plt.show()
```

```python
# Gráfico de pizza
plt.pie(contagem_idades.values, labels=contagem_idades.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Gráfico de Pizza das Idades')
plt.show()
```
